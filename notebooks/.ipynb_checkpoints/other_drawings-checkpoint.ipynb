```python
% matplotlib inline
import matplotlib.pyplot as plt
import numpy as np

fig = plt.figure()
ax = fig.add_subplot(111)

t = np.arange(0, 100, 1)
t1 = np.arange(0, 20, 1)
t2 = np.arange(20, 60, 1)
t3 = np.arange(60, 80, 1)
t4 = np.arange(80, 100, 1)
s = np.concatenate( [ 0 * t1, 1.0 / 40 * (t2 - 20),  1 - 1.0 / 20 * (t3 - 60), 0 * t4 ])

ax.plot(t, s)

ax.annotate('event climax', xy=(60, 1), xytext=(80, 1.2),
            arrowprops=dict(facecolor='black', shrink=0.05, width = 1),
            )

plt.axhline(linewidth=8, color='#2ca02c', xmin=0.2, xmax=0.6)
plt.axhline(linewidth=8, color='#d62728', xmin=0.6, xmax=0.8)

plt.xlabel('time (frames)')
plt.ylabel('progres')
plt.title('Event progress according to time')
plt.grid(True)
plt.savefig("event_progress.png")
plt.show()

```
